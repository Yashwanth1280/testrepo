
### Task 1: Exploring Relationships with Scatter Plots



#### 1. Scatter Plot for Age vs. Job Satisfaction



Visualize the relationship between respondents' age (`Age`) and job satisfaction (`JobSatPoints_6`). Use this plot to identify any patterns or trends.




```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Connect to the SQLite database
conn = sqlite3.connect('survey-results-public.sqlite')

# --- Task: Scatter Plot for Age vs. Job Satisfaction ---

# Define the SQL query to select 'Age' and 'JobSatPoints_6'
QUERY = "SELECT Age, JobSatPoints_6 FROM main"

# Execute the query and load the result into a pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# --- Data Cleaning and Preparation ---
# Drop rows where either column is missing
df.dropna(subset=['Age', 'JobSatPoints_6'], inplace=True)

# Convert the 'Age' column to approximate numeric values for plotting
# We map each age range string to its approximate midpoint
age_map = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}
df['Age_Numeric'] = df['Age'].map(age_map)

# Drop any rows that might not have been mapped correctly
df.dropna(subset=['Age_Numeric'], inplace=True)


# --- Visualization with a Scatter Plot ---
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Age_Numeric', y='JobSatPoints_6', data=df, alpha=0.5)

# Add titles and labels for clarity
plt.title('Relationship Between Age and Job Satisfaction')
plt.xlabel('Approximate Age')
plt.ylabel('Job Satisfaction (JobSatPoints_6)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Close the database connection
conn.close()
```

### Task 2: Enhancing Scatter Plots


#### 1. Scatter Plot with Trend Line for Age vs. Job Satisfaction



Add a regression line to the scatter plot of Age vs. JobSatPoints_6 to highlight trends in the data.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Connect to the SQLite database
conn = sqlite3.connect('survey-results-public.sqlite')

# --- Task: Scatter Plot with Trend Line for Age vs. Job Satisfaction ---

# Define the SQL query to select 'Age' and 'JobSatPoints_6'
QUERY = "SELECT Age, JobSatPoints_6 FROM main"

# Execute the query and load the result into a pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# --- Data Cleaning and Preparation ---
# Drop rows where either column is missing
df.dropna(subset=['Age', 'JobSatPoints_6'], inplace=True)

# Convert the 'Age' column to approximate numeric values for plotting
# We map each age range string to its approximate midpoint
age_map = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}
df['Age_Numeric'] = df['Age'].map(age_map)

# Drop any rows that might not have been mapped correctly
df.dropna(subset=['Age_Numeric'], inplace=True)


# --- Visualization with a Scatter Plot and Regression Line ---
# Use seaborn's regplot to create a scatter plot with a linear regression line
plt.figure(figsize=(12, 8))
sns.regplot(x='Age_Numeric', y='JobSatPoints_6', data=df,
            scatter_kws={'alpha':0.3}, line_kws={'color':'red'})

# Add titles and labels for clarity
plt.title('Trend of Job Satisfaction with Age')
plt.xlabel('Approximate Age')
plt.ylabel('Job Satisfaction (JobSatPoints_6)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Close the database connection
conn.close()

```

#### 2. Scatter Plot for Age vs. Work Experience


Visualize the relationship between Age (`Age`) and Work Experience (`YearsCodePro`) using a scatter plot.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Connect to the SQLite database
conn = sqlite3.connect('survey-results-public.sqlite')

# --- Task: Scatter Plot for Age vs. Work Experience ---

# Define the SQL query to select 'Age' and 'YearsCodePro'
QUERY = "SELECT Age, YearsCodePro FROM main"

# Execute the query and load the result into a pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# --- Data Cleaning and Preparation ---
# Drop rows where either column is missing
df.dropna(subset=['Age', 'YearsCodePro'], inplace=True)

# Convert the 'Age' column to approximate numeric values
age_map = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}
df['Age_Numeric'] = df['Age'].map(age_map)

# Convert 'YearsCodePro' to a numeric type, handling special strings
df['YearsCodePro'] = df['YearsCodePro'].replace({'Less than 1 year': 0.5, 'More than 50 years': 51})
df['YearsCodePro_Numeric'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Drop any rows that failed conversion
df.dropna(subset=['Age_Numeric', 'YearsCodePro_Numeric'], inplace=True)


# --- Visualization with a Scatter Plot ---
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Age_Numeric', y='YearsCodePro_Numeric', data=df, alpha=0.5)

# Add titles and labels for clarity
plt.title('Relationship Between Age and Professional Coding Experience')
plt.xlabel('Approximate Age')
plt.ylabel('Years of Professional Coding Experience')
plt.grid(True)
plt.tight_layout()
plt.show()

# Close the database connection
conn.close()

```

### Task 3: Combining Scatter Plots with Additional Features


#### 1. Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size



Create a bubble plot to explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), with bubble size representing age.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Connect to the SQLite database
conn = sqlite3.connect('survey-results-public.sqlite')

# --- Task: Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size ---

# Define the SQL query to select the necessary columns
QUERY = "SELECT ConvertedCompYearly, JobSatPoints_6, Age FROM main"

# Execute the query and load the result into a pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# --- Data Cleaning and Preparation ---
# Drop rows where any of the critical columns are missing
df.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6', 'Age'], inplace=True)

# Convert the 'Age' column to approximate numeric values for bubble size
age_map = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}
df['Age_Numeric'] = df['Age'].map(age_map)
df.dropna(subset=['Age_Numeric'], inplace=True)

# Filter out extreme salary outliers for a more meaningful plot
df_filtered = df[df['ConvertedCompYearly'] <= 300000]

# --- Visualization with a Bubble Plot ---
plt.figure(figsize=(14, 10))
sns.scatterplot(data=df_filtered,
                x='ConvertedCompYearly',
                y='JobSatPoints_6',
                size='Age_Numeric',      # Use Age for bubble size
                sizes=(20, 500),         # Define the range of bubble sizes
                alpha=0.6,               # Use transparency for overlapping points
                hue='Age_Numeric',       # Color bubbles by age for better distinction
                palette='viridis')

# Add titles and labels for clarity
plt.title('Compensation vs. Job Satisfaction (Bubble Size by Age)')
plt.xlabel('Annual Compensation (USD, up to $300k)')
plt.ylabel('Job Satisfaction (JobSatPoints_6)')
plt.legend(title='Age')
plt.grid(True)
plt.tight_layout()
plt.show()

# Close the database connection
conn.close()

```

#### 2. Scatter Plot for Popular Programming Languages by Job Satisfaction


Visualize the popularity of programming languages (`LanguageHaveWorkedWith`) against job satisfaction using a scatter plot. Use points to represent satisfaction levels for each language.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Connect to the SQLite database
conn = sqlite3.connect('survey-results-public.sqlite')

# --- Task: Scatter Plot for Programming Languages by Job Satisfaction ---

# Define the SQL query to select the necessary columns
QUERY = "SELECT LanguageHaveWorkedWith, JobSatPoints_6 FROM main"

# Execute the query and load the result into a pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# --- Data Cleaning and Preparation ---
# Drop rows where either of the critical columns is missing
df.dropna(subset=['LanguageHaveWorkedWith', 'JobSatPoints_6'], inplace=True)

# Process the programming language data
# Split the semicolon-separated strings into lists and explode them so each language gets its own row
df_langs = df.assign(LanguageHaveWorkedWith=df['LanguageHaveWorkedWith'].str.split(';')).explode('LanguageHaveWorkedWith')

# --- Aggregate Data for Plotting ---
# Calculate the median job satisfaction and the count for each language
lang_stats = df_langs.groupby('LanguageHaveWorkedWith')['JobSatPoints_6'].agg(['median', 'count']).reset_index()

# For better visualization, let's filter for languages with a significant number of users (e.g., more than 1000)
lang_stats_filtered = lang_stats[lang_stats['count'] > 1000]


# --- Visualization with a Scatter Plot (Bubble Plot) ---
plt.figure(figsize=(16, 10))
sns.scatterplot(data=lang_stats_filtered,
                x='median',
                y='LanguageHaveWorkedWith',
                size='count',          # Use the count of users for bubble size
                sizes=(100, 2000),     # Define the range of bubble sizes
                alpha=0.7,
                hue='median',          # Color by median satisfaction
                palette='viridis_r')

# Add titles and labels for clarity
plt.title('Median Job Satisfaction by Programming Language (Bubble Size by Popularity)')
plt.xlabel('Median Job Satisfaction (JobSatPoints_6)')
plt.ylabel('Programming Language')
plt.legend(title='Popularity (Count)', loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(True)
plt.tight_layout()
plt.show()

# Close the database connection
conn.close()

```
