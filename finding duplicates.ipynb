## Identify and Analyze Duplicates


### Task 1: Identify Duplicate Rows
1. Count the number of duplicate rows in the dataset.
3. Display the first few duplicate rows to understand their structure.


```python
import pandas as pd
import numpy as np

# The URL for the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/VYPrOu0Vs3I0hKLLjiPGrA/survey-data-with-duplicate.csv"

# Load the dataset into a pandas DataFrame
df = pd.read_csv(file_path)

# Display the initial dimensions of the dataframe
print("Initial DataFrame shape (rows, columns):", df.shape)

# --- A. Identify and remove duplicate rows ---

# Find the number of duplicate rows
num_duplicates = df.duplicated().sum()
print(f"\nFound {num_duplicates} duplicate rows.")

# Display the first few duplicate rows to understand their structure
print("\nDisplaying first 5 duplicate rows:")
# The keep='first' argument marks all subsequent duplicates as True
duplicate_rows = df[df.duplicated()]
print(duplicate_rows.head())


# Remove duplicate rows from the DataFrame
df.drop_duplicates(inplace=True)

# Display the new dimensions of the dataframe to confirm removal
print("\nDataFrame shape after removing duplicates:", df.shape)

```

### Task 2: Analyze Characteristics of Duplicates
1. Identify duplicate rows based on selected columns such as MainBranch, Employment, and RemoteWork. Analyse which columns frequently contain identical values within these duplicate rows.
2. Analyse the characteristics of rows that are duplicates based on a subset of columns, such as MainBranch, Employment, and RemoteWork. Determine which columns frequently have identical values across these rows.
   


```python
import pandas as pd

# The URL for the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/VYPrOu0Vs3I0hKLLjiPGrA/survey-data-with-duplicate.csv"

# Load the dataset into a pandas DataFrame
df = pd.read_csv(file_path)


# --- Task 1: Identify and Display Duplicate Rows ---

print("--- Analyzing Duplicates Based on ALL Columns ---")
# Count the number of duplicate rows in the dataset (based on all columns)
num_duplicates_all_cols = df.duplicated().sum()
print(f"Number of duplicate rows found: {num_duplicates_all_cols}")

# Display the first few duplicate rows to understand their structure
print("\nFirst 5 duplicate rows:")
print(df[df.duplicated()].head())


# --- Task 2: Analyze Characteristics of Duplicates based on a subset of columns ---

print("\n\n--- Analyzing Duplicates Based on a SUBSET of Columns ---")
# Define the subset of columns to check for duplicates
subset_cols = ['MainBranch', 'Employment', 'RemoteWork']

# Find all rows that are duplicates based on the subset of columns
# keep=False marks all occurrences of duplicates, which is useful for analysis
subset_duplicates = df[df.duplicated(subset=subset_cols, keep=False)].copy()

print(f"Found {len(subset_duplicates)} rows that are duplicates based on {subset_cols}.")

# Analyze which columns frequently contain identical values within these duplicate rows
# A low uniqueness ratio indicates that a column has few unique values, meaning it's often identical.
if not subset_duplicates.empty:
    uniqueness_ratio = subset_duplicates.nunique() / len(subset_duplicates)
    print("\nTop 15 columns that most frequently have identical values within these duplicate rows:")
    print(uniqueness_ratio.sort_values().head(15))
else:
    print("\nNo duplicate rows found based on the specified subset of columns.")
```

### Task 4: Strategic Removal of Duplicates
1. Decide which columns are critical for defining uniqueness in the dataset.
2. Remove duplicates based on a subset of columns if complete row duplication is not a good criterion.


```python
import pandas as pd

# The URL for the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/VYPrOu0Vs3I0hKLLjiPGrA/survey-data-with-duplicate.csv"

# Load the dataset into a pandas DataFrame
df = pd.read_csv(file_path)

print("--- Task 4: Strategic Removal of Duplicates ---")

# Display the initial dimensions of the dataframe
print(f"Initial DataFrame shape: {df.shape}")

# --- Decide which columns are critical for defining uniqueness ---
# For a survey, the 'ResponseId' is the most critical column for identifying a unique response.
# If two rows have the same 'ResponseId', it's a clear case of a duplicate entry.
critical_columns = ['ResponseId']
print(f"\nDefining uniqueness based on the following critical column(s): {critical_columns}")

# --- Remove duplicates based on the subset of critical columns ---
# We check for duplicates in the 'ResponseId' column.
num_duplicates_subset = df.duplicated(subset=critical_columns).sum()
print(f"Found {num_duplicates_subset} duplicate rows based on the critical column(s).")

# Remove these duplicates, keeping the first occurrence
df.drop_duplicates(subset=critical_columns, keep='first', inplace=True)

# Display the new dimensions of the dataframe to confirm removal
print(f"DataFrame shape after strategic removal of duplicates: {df.shape}")
```
